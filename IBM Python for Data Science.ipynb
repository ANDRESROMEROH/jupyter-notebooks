## Working with Tuples

**Important:** The main difference between Tuples and List is that Lists are mutables and Tuples are not!

```python
tuple = ('Apple', 'Orange', 'Watermelon')
print(tuple)
```

```python
tuple = ('Apple', 'Orange', 'Watermelon')
if 'Apple' in tuple:
    print('Yes, Apple is in the fruits tuple')
```

```python
tuple = ('Apple', 'Orange', 'Watermelon', 'Lemon', 'Avocado')
print(tuple[0:3])
print(tuple[::2])
print(tuple[1:])
```

```python
thistuple = ("Apple", "Banana", "Cherry")
print(len(thistuple)) 
```

## Working with Sets

### How to create a Set:

### Set Functions / Conditionals / Operations:

```python
list = ['Michael Jackson', 'Kurt Cobain', 'Axel Roses']
newSet = set(list)
print(newSet)
```

```python
bands = {'AC/DC', 'Guns n Roses', 'Kiss'}
if 'AC/DC' in bands:
    print('Band AC/DC found!')
```

```python
bands_set_1 = {'AC/DC', 'Guns n Roses', 'Kiss'}
bands_set_2 = {'Iron Maiden', 'Metallica', 'Guns n Roses', 'AC/DC'}
print(bands_set_1 & bands_set_2)# UNION

union = bands_set_1.union(bands_set_2)
print(union)
```

```python
bands_set = {'AC/DC', 'Guns n Roses', 'Kiss'}
bands_subset = {'AC/DC','Kiss'}

if bands_subset.issubset:
    print('Is Subset!')
```
